## IMage histograms
#### Color image histograms

```python
path1 = "D:\\Code\\Personal-Backup\\AIML\\OPENCV\\sample_images\\messi5.jpg"
img = cv2.imread(path1)
cv2.imshow("Referece",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.show()
b,g,r = cv2.split(img)
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,6)) #axes is list object.
axes[0][0].hist(b.ravel(),bins=256,range=(0,256)) # 8 bit utf, 2**8 bit representation 0, 256(excluded) and bins is the curve
axes[0][0].set_title("Blue histogram")

axes[0][1].hist(g.ravel(),bins=256,range=(0,256))
axes[0][1].set_title("Green Histogram")

axes[1][0].hist(r.ravel(),bins=256,range=(0,256))
axes[1][0].set_title("Red histogram")
fig.tight_layout()

```

## Thresholding

```python
path1 = "D:\\Code\\Personal-Backup\\AIML\\OPENCV\\sample_images\\lena.jpg"
img = cv2.imread(path1,0)
cv2.imshow('leena',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret,thresh1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY) #100-lower bound of threshold, 255-upper bound of threshold
ret, thresh2 = cv2.threshold(img,100,255,cv2.THRESH_BINARY_INV)
cv2.imshow("Thresh hold image",thresh1)
cv2.imshow("inverse Thresh hold image",thresh2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

###### Supported Parameters for Threshold Function<br>cv2.THRESH_BINARY<br>cv2.THRESH_BINARY_INV<br>cv2.THRESH_TRUNC<br>cv2.THRESH_TOZERO<br>cv2.THRESH_TOZERO_INV

### OSTU binarization

```python
path1 = "D:\\Code\\Personal-Backup\\AIML\\OPENCV\\sample_images\\apple.jpg"
img = cv2.imread(path1,0) #0-channel
cv2.imshow("aple",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot()
plt.hist(img.ravel(),bins=256,range=(0,256))
plt.show()
```

```python
ret, thresh1 = cv2.threshold(img,110,255,cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img,110,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print(ret)
cv2.imshow("threshold images",thresh1)
cv2.imshow("ostu threshold images",thresh2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Adaptive thresholding

```python
path1 = "D:\\Code\\Personal-Backup\\AIML\\OPENCV\\sample_images\\sudoku.png"
img = cv2.imread(path1,0) # work only with binary image
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #simple thresholding
cv2.imshow("global thresh holding",th1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,4)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,4)             
cv2.imshow("Adaptive Mean Threshold",th2)
cv2.imshow("Adaptive Gaussian Threshold",th3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
