```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.show()
```

```python
data = pd.read_csv("car_sales.csv")
pd.set_option('display.max_columns', None) #to display all the columns
# pd.set_option('display.min_rows',None) # display the rows
data.head() #looking at first five rows

```

```python
#plooting horse Power vs.  Milege on highway
fig, draw = plt.subplots()
draw.plot(data['Horsepower'],data['MPG.highway'],'r--')
draw.set_xlabel('HorsePower')
draw.set_ylabel('MPG.Highway')
draw.set_title("MilegeOnHighway v/s Horesepower")
```

```python
#for proper plotting we have to short the data
sort_data = data.sort_values('Horsepower',ascending=True)
data.head()
```

```python
fig, draw = plt.subplots()
draw.plot(sort_data['Horsepower'],sort_data['MPG.highway'],'r--')
draw.set_xlabel('HorsePower')
draw.set_ylabel('MPG.Highway')
draw.set_title("MilegeOnHighway v/s Horesepower")
```

```python
#floating circle type marker , all parameter is sAMe as above.
#using unsorted data to plotting
fig, flag = plt.subplots()
flag.plot(data['Horsepower'],data['MPG.highway'],'ro')
flag.set_xlabel("Horsepower",fontsize=13)
flag.set_ylabel("Mileage on Highway",fontsize=13)
flag.set_title("Horsepower vs. Mileage",fontsize=15)
```

```python
#plotting with sorted data. Scatter plot
fig,axes=plt.subplots()
axes.scatter(sort_data["Horsepower"],sort_data["MPG.highway"],c="g")
axes.set_xlabel("Horsepower",fontsize=13)
axes.set_ylabel("Mileage on Highway",fontsize=13)
axes.set_title("Horsepower vs. Mileage",fontsize=15)
```

### Bar graph

```python
fig, pict = plt.subplots()
pict.bar(sort_data['Passengers'],sort_data['Passengers'].value_counts()[sort_data['Passengers']],color='yellow',edgecolor='red' )
```
