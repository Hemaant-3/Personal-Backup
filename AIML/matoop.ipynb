#### Matplot object oriented API

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.show() # for showing figures
```

```python
#sample data
x = np.arange(0,20)
y = x**2
```

```python
# plotiing data
fig = plt.figure()  # Creatinh a blank figures
# [left,bottom,width,height] of the new axes
my_chart = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # adding axes to figure. and define my own chart my_chart
# all quantities are in fractions of figure width and height.
my_chart.plot(x, y)  # plotting on the defined axes by me.
my_chart.set_xlabel("X-Axis")
my_chart.set_ylabel("Y-Axis")
my_chart.set_title("Square")
```

#### Multiple plots in one figure using custom axes

```python
fig = plt.figure()
plot_1 = fig.add_axes([0,0,0.5,0.5])
plot_2= fig.add_axes([1,0,2.2,2.2]) #adding axes

plot_1.plot(x,y,'ro')
plot_1.set_xlabel("X-Axis")
plot_1.set_ylabel("Y-Axis")
plot_1.set_title("Fig. 'A'")

plot_2.plot(x,y,'go')
plot_2.set_xlabel("X-Axis")
plot_2.set_ylabel("Y-Axis")
plot_2.set_title("Fig. 'B'")
fig.tight_layout()
```

```python
#Axes within axes
fig = plt.figure()
my_fig1 = fig.add_axes([0,0,0.9,0.9])
my_fig2 = fig.add_axes([0.15,0.5,0.3,0.3])

my_fig1.plot(x,y)
my_fig2.plot(x,y)
```

### Multiplot using sub_plot

```python
# fig, axes_sub = plt.subplots(nrows=2,ncols=2) #axes is a list object
fig, axes_sub = plt.subplots(nrows=2, ncols=2,figsize=(7,5), dpi=150) #adding custom size and dpi
axes_sub[0][0].plot(x,y,'g')
axes_sub[0][1].plot(x,y,'y')
axes_sub[1][1].plot(x,y,'r')
axes_sub[1][0].plot(x,y,'b')
fig.subplots_adjust(hspace=0.6,wspace=0.6)
```

```python
#Sample plotting in one figure
fig,plot_3 = plt.subplots(nrows=1,ncols=1)
plot_3.plot(x,y,'r')
plot_3.plot(y,x,'g')
plot_3.set_xlabel("X-Axis")
plot_3.set_ylabel("Y-Axis")
plot_3.set_title("Square")
```

### LEgend and legend position

```python
fig, polt = plt.subplots(nrows=1,ncols=1)
polt.plot(x,y,'g',label="plot1")
polt.plot(y,x,'r',label="plot2")
# polt.legend(loc=0)
polt.legend(loc=4)
# polt.legend() # show to table.

```

```python
fig, polt = plt.subplots(nrows=1,ncols=1)
polt.plot(x,y,'g',label="plot1")
polt.plot(y,x,'r',label="plot2")
## Axis limit.
polt.set_xlim(0,500)
polt.set_ylim(0,200)
polt.legend()
```

```python

fig, polt = plt.subplots(nrows=1,ncols=1)
polt.plot(x,y,'g',label="plot1")
polt.plot(y,x,'r',label="plot2")
## Adding, tickers, padding,
polt.xaxis.labelpad = 10
polt.yaxis.labelpad = 20
polt.set_xticks(np.arange(0,400,30))
polt.set_yticks(np.arange(0,460,30))
polt.legend()
#saving file 
fig.savefig("Newfig",dpi=300)
```

```python
#Enabling Grid
fig, ploot = plt.subplots(nrows=1,ncols=1)
ploot.plot(x,y,'go')
ploot.plot(y,x,'ro')
ploot.set_xlabel('X-Axis')
ploot.set_ylabel('Y-Axis')
ploot.set_title('Square')
ploot.grid(True)
```

### Anatomy of figure and spines

```python
fig,ploot = plt.subplots(nrows=1,ncols=1)
ploot.plot(x,y,'ro')
ploot.plot(y,x,'go')
ploot.set_xlabel('X-Axis')
ploot.set_ylabel('Y-Axis')
ploot.set_title('Square')
ploot.grid(True)
ploot.spines['left'].set_color('Yellow') #change the color of specefied color
ploot.spines['left'].set_linewidth(3)
ploot.spines['right'].set_color("Green")
ploot.spines['right'].set_linewidth(4)
ploot.spines['top'].set_color('Red')
ploot.spines['top'].set_linewidth(5)
ploot.spines['bottom'].set_color('Blue')
ploot.spines['bottom'].set_linewidth(5)
```

```python
from matplotlib.ticker import MultipleLocator
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.plot(y,x,"g")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.xaxis.set_major_locator(MultipleLocator(100))  # set interval of major grid line
axes.xaxis.set_minor_locator(MultipleLocator(20))   # set interval of minor grid line
axes.grid(True,which='major',color="pink")          # enable minor grids
axes.grid(True,which='minor',color="black")         # enable major grids
fig.savefig("Minorfig.jpeg",dpi=250)
```

```python
fig,draw=plt.subplots(nrows=1,ncols=1)
draw.plot(x,y,'r--')
draw.set_xlabel("X-axis")
draw.set_ylabel("Y-Axis")
draw.text(9,170,'$y=x^2$',fontsize=20,color='red')
```
